```python
import matplotlib.pyplot as plt
squares = [1,4,9,16,25]
plt.plot(squares)
plt.show
```

```python
# add thickness
plt.plot(squares, linewidth = 6)
# ?set title and label axis
plt.title('Squared numbers', fontsize= 20)
plt.xlabel('Value', fontsize = 10)
plt.ylabel('Square of value', fontsize = 10)
plt.show
```

```python
input_values = [1,2,3,4,5]
plt.plot(input_values, squares, linewidth = 5)
```

```python
# scatter plot
plt.scatter(2,4, s= 400)
plt.title('Squared Number', fontsize = 24)
plt.xlabel('Value', fontsize = 10)
plt.ylabel('Squares', fontsize = 10)
# add tick label
plt.tick_params(axis='both', which='major', labelsize=14)
plt.show()
```

```python
xvalues = [1,2,3,4,5]
yvalues = [1,4,9,16,25]

plt.scatter(xvalues, yvalues, linewidths=2)
```

```python
x_values = list(range(1, 101))
y_value = [x**2 for x in x_values]

plt.scatter(x_values, y_value,  linewidths=2, edgecolors= 'none', s =40)
```

```python
x_values = list(range(1, 101))
y_value = [x**2 for x in x_values]

plt.scatter(x_values, y_value,linewidths=2, edgecolors= 'none', s =40, c= y_value, cmap=plt.cm.Blues)
```

```python
x_axis = [1,2,3,4,5]
y_axis = [1,8,27,64,125] 
plt.scatter(x_axis, y_axis, linewidths=1, s= 30)
```

```python
x_axis = list(range(0, 500))
y_axis = [x**3 for x in x_axis]
plt.scatter(x_axis, y_axis, linewidths=2, c = y_axis, cmap=plt.cm.Blues)
plt.title('Cube', fontsize = 24)
plt.xlabel('Values', fontsize = 10)
plt.ylabel('Cubes', fontsize = 10)
plt.show()
```
