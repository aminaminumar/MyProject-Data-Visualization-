```python
from random import choice

class RandomWalk():
    # A class to generate ramdom walks
    def __init__(self, num_points = 5000):
        self.num_points = num_points
        
        # all work starts at zero
        self.x_values = [0]
        self.y_values = [0]

    def fill_walk(self):
        """calculate al the points in the walk"""

        # keep walking until the result reach the desired 
        while len(self.x_values) < self.num_points:
            # decide which direction to go
            x_direction = choice([1, -1])
            x_distance = choice([0,1,2,3,4])
            x_steps = x_direction * x_distance

            y_direction = choice([1, -1])
            y_distance = choice([0,1,2,3,4])
            y_steps = y_direction * y_distance

            # reject move that go nowhere
            if x_steps == 0 and y_steps == 0:
                continue

            # calculate the next step
            x_next = self.x_values[-1] + x_steps
            y_next = self.y_values[-1] + y_steps

            self.x_values.append(x_next)
            self.y_values.append(y_next)




```

```python
import matplotlib.pyplot as plt

rw = RandomWalk()
rw.fill_walk()
plt.scatter(rw.x_values, rw.y_values, s=15)
plt.show()
```

```python
class RandomWalk_modified():
    # A class to generate ramdom walks
    def __init__(self, num_points = 5000):
        self.num_points = num_points
        
        # all work starts at zero
        self.x_values = [0]
        self.y_values = [0]

    def get_step(self):
            # get the direction and distance to find the steps
            direction = choice([-1, 0, 1])
            distance = choice([0,1,2,3,4])
            steps = distance * direction
            return steps
        
    def fill_walk2(self):
        """calculate all the points in the walk"""

        # keep walking until the result reach the desired 
        while len(self.x_values) < self.num_points:
            # decide which direction to go
            x_steps = self.get_step()
            y_steps = self.get_step()

            # reject move that go nowhere
            if x_steps == 0 and y_steps == 0:
                continue

            # calculate the next step
            x_next = self.x_values[-1] + x_steps
            y_next = self.y_values[-1] + y_steps

            self.x_values.append(x_next)
            self.y_values.append(y_next)
```
