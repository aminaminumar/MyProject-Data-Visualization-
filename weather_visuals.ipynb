```python
# Extracting and reading data
# get tmperature from files

file_name = 'sitka_weather_data.csv'
with open(file_name) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    # print(header_row)

    temperature = []
    for row in reader:
        # conver the temperature
    
        temperature.append(int(row[1]))
    print(temperature)
```

```python
# visualize the temp using pyplot
import matplotlib.pyplot as plt

# plot data
fig = plt.figure(dpi=128, figsize=(10, 6))
plt.plot(temperature, c= 'red')
plt.title('daily Temperature')
plt.xlabel('', fontsize = 24)
plt.ylabel('Temperature', fontsize = 24)
plt.tick_params(axis= 'both', which= 'major' )
plt.show()
```

## ADDING DATETIME MODULE

```python
from datetime import datetime
# get dates and temp from file
file_name = 'sitka_weather_data.csv'
with open(file_name) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    # print(header_row)

    dates, temperature = [], []
    for row in reader:
        current_date = datetime.strptime(row[0], "%Y-%m-%d")
        dates.append(current_date)

        # conver the temperature
        temperature.append(int(row[1]))

fig = plt.figure(dpi=128, figsize=(10, 6))
plt.plot(dates, temperature, c= 'red')
plt.title('daily Temperature')
plt.xlabel('', fontsize = 24)
plt.ylabel('Temperature', fontsize = 24)
fig.autofmt_xdate()
plt.tick_params(axis= 'both', which= 'major' )
plt.show()
```

```python
# plot to show humidity
file_name = 'sitka_weather_data.csv'
with open(file_name) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    # print(header_row)

    dates, temperature, humidity = [], [], []
    for row in reader:
        current_date = datetime.strptime(row[0], "%Y-%m-%d")
        dates.append(current_date)
    
        # conver the temperature to int
        temperature.append(int(row[1]))
        humidity.append(int(row[2]))


fig = plt.figure(dpi=128, figsize=(10, 6))
plt.plot(dates, temperature, c= 'red')
plt.plot(dates, humidity, c= 'blue')
plt.title('daily Temperature')
plt.xlabel('', fontsize = 24)
plt.ylabel('Temperature', fontsize = 24)
fig.autofmt_xdate()
plt.tick_params(axis= 'both', which= 'major' )
plt.fill_between(dates, temperature, humidity, facecolor='blue', alpha=0.1)

plt.show()
```
